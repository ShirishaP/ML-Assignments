### Data preprocessing

```python
df1['data3'] = pd.Categorical(df1['data3'])
df1['data10'] = pd.Categorical(df1['data10'])
df2['data3'] = pd.Categorical(df1['data3'])
df2['data10'] = pd.Categorical(df1['data10'])

print (df1.dtypes)
```

```python
print (df1['data3'].cat.categories)
print (df1['data10'].cat.categories)
print (df2['data3'].cat.categories)
print (df2['data10'].cat.categories)
```

```python
df1=df1.drop('data10',axis=1) #drop data10 as it is not necessary
```

```python
gender = pd.get_dummies(df1['data9'],drop_first=True)
gender1 = pd.get_dummies(df2['data9'],drop_first=True)
gender
```

```python
df1 = df1.drop('data9',axis=1)
df2= df2.drop('data9',axis=1)
df1.head()
```

```python
df1=pd.concat([df1,gender],axis=1)
df2=pd.concat([df2,gender1],axis=1)
df1.head()
```

```python
#change categorical to numerical
from sklearn.preprocessing import LabelEncoder
# creating instance of labelencoder
labelencoder = LabelEncoder()
# Assigning numerical values A=0 , B=1,...
df1['data3'] = labelencoder.fit_transform(df1['data3'])
df2['data3'] = labelencoder.fit_transform(df2['data3'])
df1.head() 
```

```python
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
df1['data5']=df1['data5'].fillna(df1['data5'].dropna().mean())
df1['data6']=df1['data6'].fillna(df1['data6'].dropna().mean())
df1['data7']=df1['data7'].fillna(df1['data7'].dropna().mean())
df1['data8']=df1['data8'].fillna(df1['data8'].dropna().mean())
df1['output']=df1['output'].fillna(df1['output'].dropna().mean())
```

```python
df1.isnull().sum()
```

```python
X= df1.drop('output',axis=1)
Y = df1.loc[:,['output']].values
Y.shape
```

### Normalization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X.shape
```

```python
Y = Y.reshape(-1)
Y.shape
```

### Test Train Split

```python
#test train split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.20,random_state=0)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression() #training the model
reg = lr.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

### Mean squared error

```python
from sklearn.metrics import mean_squared_error, r2_score
print('Mean squared error: %.2f'
      % mean_squared_error(y_test,y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))
R2= r2_score(y_test, y_pred)
```
